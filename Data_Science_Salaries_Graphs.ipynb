```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ds_salaries.csv')
```

```python
# Filter out salaries where the currency is USD for simplicity
df_usd = df[df['salary_currency'] == 'USD']
```

```python
# Group data by job title and calculate average salary
avg_salary_by_job_title = df_usd.groupby('job_title')['salary'].mean().sort_values()
```

```python
avg_salary_by_job_title
```

```python
# Plot histogram
plt.figure(figsize=(10, 8))
plt.barh(avg_salary_by_job_title.index, avg_salary_by_job_title.values, color='skyblue')
plt.xlabel('Average Salary (USD)')
plt.title('Average Salary by Job Title')
plt.gca().invert_yaxis()  # Invert y-axis to display job titles from top to bottom
plt.tight_layout()
plt.show()
```

```python
#Distribution of remote work ratio
plt.figure(figsize=(8, 6))
df['remote_ratio'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Remote Work Ratio')
plt.xlabel('Remote Work Ratio')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 4. Geographic distribution of salaries
plt.figure(figsize=(12, 10))
df.groupby('employee_residence')['salary'].mean().sort_values().plot(kind='bar', color='orange')
plt.title('Average Salary by Employee Residence')
plt.xlabel('Employee Residence')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
